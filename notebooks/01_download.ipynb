<h2 style="font-family: Cambria;">
Configuration
</h2>

```python
DESTINATION_FOLDER = "/data1/wangteng/raw"

# List of Kaggle dataset slugs to download.
DATASETS_TO_DOWNLOAD = [
    "blastnet/channelflow-dns-re544-seq-p000-020",
    # "blastnet/channelflow-dns-re544-seq-p021-041",
    # "blastnet/channelflow-dns-re544-seq-p042-062",
    # "blastnet/channelflow-dns-re544-seq-p063-083",
    # Add more sequences here if needed...
]
```

<h2 style="font-family: Cambria;">
Helper Function for downloading
</h2>

```python
def download_and_unzip(dataset_slug, base_destination):
    """
    Downloads a single dataset if it does not already exist.
    
    Args:
        dataset_slug (str): The Kaggle dataset identifier (e.g., 'user/dataset').
        base_destination (str): The root directory where datasets are stored.
    """
    dataset_name = dataset_slug.split('/')[-1]
    target_dir = os.path.join(base_destination, dataset_name)

    if os.path.exists(target_dir) and len(os.listdir(target_dir)) > 0:
        print(f"[SKIP] Dataset seems to exist in '{target_dir}'. Skipping download.")
        return

    # Create the directory if it doesn't exist.
    if not os.path.exists(target_dir):
        print(f"Creating directory: {target_dir}")
        os.makedirs(target_dir, exist_ok=True)

    # Execute.
    try:
        print("Starting download via Kaggle API...")
        kaggle.api.dataset_download_files(
            dataset_slug, 
            path=target_dir, 
            unzip=True, 
            quiet=False 
        )
        print(f"[SUCCESS] Downloaded and unzipped: {dataset_slug}")  

    except Exception as e:
        print(f"[ERROR] {e}")
```

<h2 style="font-family: Cambria;">
Main Execution Loop
</h2>

```python
if not os.path.exists(DESTINATION_FOLDER):
    print(f"Creating base directory: {DESTINATION_FOLDER}")
    os.makedirs(DESTINATION_FOLDER, exist_ok=True)

print(f"Starting batch download of {len(DATASETS_TO_DOWNLOAD)} dataset(s)...\n")
for slug in DATASETS_TO_DOWNLOAD:
    download_and_unzip(slug, DESTINATION_FOLDER)

print("\nBatch download process finished.")
```
